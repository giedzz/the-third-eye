```python
import pandas as pd
import json
```

```python
input2_csv = 'polygon_data.csv'
polygon2_data_df = pd.read_csv(input2_csv)
polygon2_data_df.head(10)
```

```python
input_csv = 'polygon_data.csv'
polygon_data_df = pd.read_csv(input_csv)
polygon_data_df.head(10)
```

```python
polygon_data_df_concat = polygon_data_df.append(polygon2_data_df, ignore_index=True)
```

```python
input_csv = 'polygon_data_v2.csv'

geo_json_dict = {"type": "FeatureCollection", "features": []}
for index, row in polygon_data_df_concat.iterrows():
    polygon_string = row['st_astext']
    polygon_string_clean = polygon_string.replace('(', '').replace('POLYGON', '').replace(')', '').split(',')
    polygon_list_clean = [x.split(' ') for x in polygon_string_clean]
    polygon_list_clean = [[float(s) for s in sublist] for sublist in polygon_list_clean]
    geo_json_dict['features'].append({
        "type": "Feature",
        "properties": {},
        "geometry": {
        "type": "Polygon",
            "coordinates": [polygon_list_clean]
        }
    })
out_file = open(input_csv.replace('.csv', '.json'), "w")
json.dump(geo_json_dict, out_file, indent = 6)
out_file.close()
```

```python
polygon_set_long = []
polygon_set_lat = []
for row_key, row_data in polygon_data_df_concat.iterrows():
    polygon_string = row_data['st_astext-2']
    polygon_string_clean = polygon_string.replace('(', '').replace('POLYGON', '').replace(')', '').split(',')
    polygon_list_clean = [x.split(' ') for x in polygon_string_clean]
    polygon_list_clean = [[float(s) for s in sublist] for sublist in polygon_list_clean]
    for pol_list in polygon_list_clean:
        polygon_set_long.append(pol_list[0])
        polygon_set_lat.append(pol_list[1])
```

```python
print(max(polygon_set_long))
print(min(polygon_set_long))
print(max(polygon_set_lat))
print(min(polygon_set_lat))
```
